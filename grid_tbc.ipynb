```python
# orthogonal case
l_a = 1.0
l_b = 1.8
l_c = 1.4

lattice = np.array([[l_a, 0, 0], [0, l_b, 0], [0, 0, l_c]])

def get_system_size():
    # Get the size of the cell.
    system_size = [
        L * ANGSTROM for L in [l_a, l_b, l_c]   
    ]
    if n_grid_def == "density":
        grid_size = [
            int(L / ANGSTROM * n_grid) for L in system_size
        ]  # int always rounds down
    elif n_grid_def == "value":
        grid_size = [n_grid, n_grid, n_grid]
    grid = np.zeros(
        [grid_size[0], grid_size[1], grid_size[2], 3]
    )
    for i in range(grid_size[0]):
        for j in range(grid_size[1]):
            for k in range(grid_size[2]):
                grid[i, j, k] = (
                    i / grid_size[0] * lattice[0]
                    + lattice[0] / (2 * grid_size[0])
                    + j / grid_size[1] * lattice[1]
                    + lattice[1] / (2 * grid_size[1])
                    + k / grid_size[2] * lattice[2]
                    + lattice[2] / (2 * grid_size[2])
                )
    return grid

get_system_size()
```
